```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt


n = 100
random_numbers = np.random.uniform(0, 1, n)

# Prueba de Kolmogorov Smirnov
statistic, p_value = stats.kstest(random_numbers, 'uniform')

print(f'Estadístico KS: {statistic}')
print(f'Valor p: {p_value}')

# Interpretar resultados
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula: los datos no son uniformes.")
else:
    print("No rechazamos la hipótesis nula: los datos son uniformes.")

# Grafica
plt.hist(random_numbers, bins=10, density=True, alpha=0.6, color='g')
plt.title('Grafica de Números Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```
